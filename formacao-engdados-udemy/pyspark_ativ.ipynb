```python
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-11-openjdk-amd64"
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("BikePesquisa") \
    .getOrCreate()

spark
```

```python
from pyspark.streaming import StreamingContext
from pyspark import SparkContext

sc = spark.sparkContext

pesquisa = sc.textFile("./pesquisabike.txt")
pesquisa.take(5)

contagem = pesquisa.flatMap(lambda word: word.split(" "))
contagem = contagem.map(lambda w: (w, 1))
contagem = contagem.reduceByKey(lambda a, b: a + b)
contagem.take(5)
contagem.saveAsTextFile("resultados")
```

```python
ssc = StreamingContext(sc, 10)
#pasta a monitorar
pesquisa = ssc.textFileStream("./resultados/")
#contagem
contagem = pesquisa.flatMap(lambda palavra: palavra.split(" "))
contagem = contagem.map(lambda pal: (pal, 1))
contagem = contagem.reduceByKey(lambda a, b: a + b)
#imprime
contagem.pprint()
ssc.start()
ssc.awaitTermination()
```
