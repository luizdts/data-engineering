```python
import boto3
import io
import psycopg2

s3 = boto3.resource(
    service_name='s3',
    region_name='sa-east-1',
    aws_access_key_id='AKIA3FLD47354AP7UICM',
    aws_secret_access_key='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'
)

bucket='engdadosimages'
prefix='imagens/'

id = 0

conn = psycopg2.connect(host='database-1.cpaq02wmqhn2.us-east-2.rds.amazonaws.com',
                        database='inventario',
                        user='postgres',
                        password='postgres')
conn.autocommit = True
cur = conn.cursor()

for objects_s3 in s3.Bucket(bucket).objects.filter(Prefix=prefix):
    if objects_s3.key.endswith('jpg') or objects_s3.key.endswith('JPG'):
        filename = objects_s3.key.split('/')[1]
        id += 1
        cur.execute("insert into arquivos (idarquivo,nomearquivo) values ("+str(id)+ ",'" + filename + "')")

conn.close()
```
